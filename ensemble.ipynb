```python
import pandas as pd
```

```python
convxl = pd.read_csv('/content/drive/MyDrive/pred/convxL.csv')
efrwm = pd.read_csv('/content/drive/MyDrive/pred/efrwm.csv')
reg = pd.read_csv('/content/drive/MyDrive/pred/reg.csv')
tf = pd.read_csv('/content/drive/MyDrive/pred/tf.csv')
submission = pd.read_csv('/content/drive/MyDrive/pred/sample_submission.csv')
convxl
```

```python
label = {'1++':0, '1+': 1, '1': 2, '2': 3, '3': 4 }
```

```python
convxl['grade'] = convxl['grade'].replace(label)
efrwm['grade'] = efrwm['grade'].replace(label)
reg['grade'] = reg['grade'].replace(label)
tf['grade'] = tf['grade'].replace(label)
```

```python
b_models = (efrwm['grade'] + reg['grade'] + tf['grade']) / 3
submission['grade'] = round(b_models)
submission['grade'].plot()
```

```python
label_de = {0:'1++', 1: '1+', 2: '1', 3: '2', 4: '3' }
submission['grade'] = submission['grade'].replace(label_de)
submission
```

```python
submission.to_csv('/content/drive/MyDrive/pred/ensemble4.csv', index=False)
```
